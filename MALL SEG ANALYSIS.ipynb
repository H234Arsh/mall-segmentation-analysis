```python
mall.corr()
```

```python
sns.heatmap(mall.corr(),annot=True,cmap='coolwarm')
```

```python
clustering1 = KMeans(n_clusters=3)
```

```python
clustering1.fit(mall[['Annual Income (k$)']])
```

```python
clustering1.labels_
```

```python
mall['Income Cluster'] = clustering1.labels_
mall.head()
```

```python
mall['Income Cluster'].value_counts()
```

```python
clustering1.inertia_
```

```python
inertia_scores=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(mall[['Annual Income (k$)']])
    inertia_scores.append(kmeans.inertia_)
inertia_scores
```

```python
plt.plot(range(1,11),inertia_scores)
```

```python
mean_val = mall.groupby('Income Cluster')[['Age', 'Annual Income (k$)', 'Spending Score (1-100)']].mean()
print(mean_val)
```

```python
clustering2 = KMeans(n_clusters=5)
clustering2.fit(mall[['Annual Income (k$)','Spending Score (1-100)']])
mall['Spending and Income Cluster'] =clustering2.labels_
mall.head()
```

```python
inertia_scores2=[]
for i in range(1,11):
    kmeans2=KMeans(n_clusters=i)
    kmeans2.fit(mall[['Annual Income (k$)','Spending Score (1-100)']])
    inertia_scores2.append(kmeans2.inertia_)
plt.plot(range(1,11),inertia_scores2)
```

```python
centers =pd.DataFrame(clustering2.cluster_centers_)
centers.columns = ['x','y']
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x=centers['x'],y=centers['y'],s=100,c='black',marker='*')
sns.scatterplot(data=mall, x ='Annual Income (k$)',y='Spending Score (1-100)',hue='Spending and Income Cluster',palette='tab10')
plt.savefig('clustering_bivaraiate.png')
```

```python
pd.crosstab(mall['Spending and Income Cluster'],mall['Gender'],normalize='index')
```

```python
mean_sp= mall.groupby('Spending and Income Cluster')[['Age', 'Annual Income (k$)','Spending Score (1-100)']].mean()
print(mean_sp)
```

```python
from sklearn.preprocessing import StandardScaler
scale = StandardScaler()
mall.head()
```

```python
dff = pd.get_dummies(mall,drop_first=True)
dff.head()
```

```python
dff = dff[['Age', 'Annual Income (k$)', 'Spending Score (1-100)','Gender_Male']]
dff.head()
```

```python
dff = scale.fit_transform(dff)
dff = pd.DataFrame(scale.fit_transform(dff))
dff.head()
```

```python
inertia_scores3=[]
for i in range(1,11):
    kmeans3=KMeans(n_clusters=i)
    kmeans3.fit(dff)
    inertia_scores3.append(kmeans3.inertia_)
plt.plot(range(1,11),inertia_scores3)
```
